```python
from google.colab import files
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt

# 파일 업로드
uploaded = files.upload()

# 업로드된 이미지 파일을 열어서 출력
for filename in uploaded.keys():
    img = Image.open(filename)

    # 이미지를 numpy 배열로 변환 (픽셀 값 확인용)
    img_np = np.array(img)

    # 픽셀 값 출력 (상위 10x10 부분만)
    print("상위 10x10 픽셀 값:")
    print(img_np[:10, :10])

    # 가우시안 블러 적용 (OpenCV를 사용)
    img_blurred = cv2.GaussianBlur(img_np, (15, 15), 0)

    # 원본 이미지와 블러 처리된 이미지 출력
    plt.figure(figsize=(10, 5))

    # 원본 이미지
    plt.subplot(1, 2, 1)
    plt.imshow(img)
    plt.title('Original Image')
    plt.axis('off')

    # 가우시안 블러 처리된 이미지
    plt.subplot(1, 2, 2)
    plt.imshow(img_blurred)
    plt.title('Gaussian Blurred Image')
    plt.axis('off')

    plt.show()
```
